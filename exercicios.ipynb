```python
inicio = int(input("Digite o primeiro termo: "))
final = int(input("Digite o número final: "))
razao = int(input("Digite a razão da PG: "))
print("Progressão Geométrica:")

termo = inicio
for i in range(inicio, final + 1):
    if termo <= final:
        proximo = termo * razao
        
        if proximo > final:
            print(termo, end=".")
        else:
            print(termo, end=",")
            
        termo = proximo
    else:
        break
```

```python
livros = []

# Definindo quantos livros o usuário quer adicionar
quantidade = int(input("Quantos livros deseja cadastrar? "))

for i in range(quantidade):
    print(f"\nCadastro do livro {i+1}:")
    titulo = input("Título: ")
    autor = input("Autor: ")
    ano = input("Ano: ")
    
    # Criando o dicionário e adicionando à lista
    novo_livro = {
        "Título": titulo,
        "Autor": autor,
        "Ano": ano
    }
    livros.append(novo_livro)

print("\nInformações dos livros: ")

cont=1
for livro in livros:
    print(f"Título{cont}: {livro['Título']}, Autor: {livro['Autor']}, Ano: {livro['Ano']}")
    cont+=1

livros.pop(0)

print("\nLivros Atualizados!!!")

cont=1
for livro in livros:
    print(f"Título{cont}: {livro['Título']}, Autor: {livro['Autor']}, Ano: {livro['Ano']}")
    cont+=1
```
