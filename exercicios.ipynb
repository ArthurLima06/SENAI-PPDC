```python
inicio = int(input("Digite o primeiro termo: "))
final = int(input("Digite o número final: "))
razao = int(input("Digite a razão da PG: "))
print("Progressão Geométrica:")

termo = inicio
for i in range(inicio, final + 1):
    if termo <= final:
        proximo = termo * razao
        
        if proximo > final:
            print(termo, end=".")
        else:
            print(termo, end=",")
            
        termo = proximo
    else:
        break
```

```python
# Criando uma nova coluna no DataFrame com a amplitude por grupo, o .transform() transforma a coluna no resultado da conta que você coloca no lambda
df['amplitude'] = df.groupby('grape_type')['volume_hl'].transform(lambda x: x.max() - x.min())

# Agora você pode ver o grape_type e a amplitude lado a lado
print(df[['grape_type', 'amplitude']].drop_duplicates())
```

```python
#primeiro pega a amostra de 25%
amostra= df.sample(frac=0.25, random_state=42)

#agrupa a coluna pela amostra e coloca na variavel grupo_amostra
grupo_amostra= amostra.groupby("grape_type")

#e faz todas as contas com essa variavel que esta ligada com a amostra e o dataframe
amp_abs_amostra= grupo_amostra["volume_hl"].max() - grupo_amostra["volume_hl"].min()
media_amostra= grupo_amostra["volume_hl"].mean()
amp_rel_amostra= (amp_abs_amostra/media_amostra)*100
variancia_amostra= grupo_amostra["volume_hl"].var(ddof=0)
desvio_amostra= grupo_amostra["volume_hl"].std(ddof=0)

#para imprimir monta um novo DataFrame com esse comando do próprio pandas
resultado_amostra= pd.DataFrame({
    "Amplitude Absoluta": amp_abs_amostra,
    "Amplitude Realtiva": amp_rel_amostra,
    "Variancia": variancia_amostra,
    "Desvio Padrão": desvio_amostra
})

print(resultado_amostra.round(2))
```

```python
import pandas as pd 

df= pd.read_csv("vinhos.csv")

#O IQR serve para dar uma noção da discrepância dos dados, quantio maior o resultado, maior a discrepância.
q1= df["value_usd"].quantile(0.25)
q3= df["value_usd"].quantile(0.75)
iqr= q3-q1

#E os limites servem para dar essa noção mas do teto e do piso dos dados.
limite_inferior= q1 - 1.5 * iqr
limite_superior= q3 + 1.5 * iqr

print("-" * 32)
print("O IQR desta coluna é: ", iqr.round(2))
print("O limite inferior é: ", limite_inferior.round(2))
print("O limite superior é: ", limite_superior.round(2))
print("-" * 32)
```

```python
import pandas as pd 

df= pd.read_csv("vinhos.csv")

#Esse comando serve para pegar o zcore, ou seja, o quanto o valor que você esta analisamdo esta distante da média.
df["zscore"]= ((df["value_usd"]-df["value_usd"].mean())/df["value_usd"].std(ddof=0)).round(2)
df.head()
```

```python
import pandas as pd 

df= pd.read_csv("vinhos.csv")

grupo= df.groupby("grape_type")

q1= grupo["value_usd"].quantile(0.25)
q3= grupo["value_usd"].quantile(0.75)
iqr= q3-q1

limite_inferior= q1 - 1.5 * iqr
limite_superior= q3 + 1.5 * iqr

#O | é o operador lógico OU, para mostrar os dois valores das duas linhas deste DataFrame 
# o map define que os valores vão ser maiores ou menores que os calculos feitos dentro das variaveis de limites
outliers_iqr= df[
    (df["volume_hl"]<df["grape_type"].map(limite_inferior))|
    (df["volume_hl"]>df["grape_type"].map(limite_superior))
]
print("-" * 30)
print(limite_inferior)
print("-" * 30)
print(limite_superior)
print("-" * 30)
print(outliers_iqr)
print("-" * 30)
```

```python
import pandas as pd 

df= pd.read_csv("vinhos.csv")

media= df.groupby("grape_type")["volume_hl"].transform("mean")
std= df.groupby("grape_type")["volume_hl"].transform(lambda x: x.std(ddof=0))
df["zscore"]= (df["volume_hl"]-media)/std

#A lógica do >/<3 serve para mostrar os números que estçao muito acima da média ou muit abaixo da média
#O | é o operador lógico OU, para mostrar os dois valores das duas linhas deste DataFrame 
outliers_zscore= df[
    (df["zscore"]>3)|
    (df["zscore"]<3)
]
amostra= df.sample(frac=0.25, random_state=42)

print("-" * 30)
print(media)
print("-" * 30)
print(std)
print("-" * 30)
print(outliers_zscore)
print("-" * 30)
print(amostra)
print("-" * 30)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 

df= pd.read_csv("vinhos.csv")

#configurar tamanho por polegada, altura e largura
plt.figure(figsize=(6,4))
#tipo de gráfico, no caso o de barras
sns.barplot(
    data= df, #base de dados para o gráfico
    x= "category", #os dados que vão estar no eixo x
    y= "value_usd",#dados do eixo y
    estimator= "mean" #cálculo quesera usado, no caso a média de valor dos vinhos
)
plt.title("preço medio por categoria") #para dar um título para o gráfico todo
plt.xlabel("categoria") #nome do eixo x
plt.ylabel("preço medio") #nome do eixo y
plt.show() #imprimir o gráfico
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 

df= pd.read_csv("vinhos.csv")

plt.figure(figsize=(6, 4))
sns.countplot( #tipo do gráfico, neste caso o de barras na horizontal.
    data= df,
    y= "grape_type",
    order= df["grape_type"].value_counts().index, #comando para fazer a contagem por tipo de uva/tabela.
    palette= "Greens_d" #define a paleta de cores.
)
plt.title("Contagem por tipo de uva")
plt.xlabel("Contagem")
plt.ylabel("Tipo de uva")
plt.show()
```

```python
#comando para agrupar por média e criar a nova tabela que será analisada no gráfico do DataFrame.
df_year= df.groupby("year")["value_usd"].mean().reset_index() 
plt.figure(figsize=(5, 3))
sns.lineplot(
    data= df_year,
    x= "year",
    y= "value_usd", 
    marker= "o", #comando para colocar bolinhas nos ápices.
    color= "crimson" 
)
plt.title("Preço médio por ano")
plt.xlabel("Ano")
plt.ylabel("Preço médio")
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.histplot(
    df["value_usd"],
    bins= 30, #isto define o intervalo, quanto maior mais detalhado.
    kde= False, #Serve para mostrar a linha que ressalata as ocilações.
    color= "steelblue"
)
plt.title("Histograma de preços")
plt.xlabel("Preço")
plt.ylabel("Frequência")
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.scatterplot(
    data= df[df["category"]=="Sales"],
    x= "volume_hl",
    y= "value_usd",
    hue= "category",
    alpha= 0.7,
    palette= "tab10"
)
plt.title("Preço vs Volume(HL)")
plt.xlabel("Volume(HL)")
plt.ylabel("Preço")
plt.legend(title="Categoria")
plt.show()
```
