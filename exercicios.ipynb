```python
produtos = {
    "Mouse":98.75,
    "Teclado":125.65,
    "Monitor":134.78,
    "Gabinete":170.00,
    "HD Externo":510.50,
    "Headset":125.45}
while True:
  produto= input("Informe o produto a pesquisar o preço ou fim para sair: ")
  if produto == "fim":
    break
  if produto in produtos:
    print(f"Produto {produto} custa {produtos[produto]}$.")
  else:
    print(f"Produto {produto} não encontrado.")
```

```python
produtos= {
    "Fogão":1200.00,
    "Frigideira":500.00,
    "Panela de pressão":1000.00
}

produtos.pop("Fogão")
#comando para remover chave e valor de um dicionário (.pop()).
for chave, valor in produtos.items():
    print(f"Valor do produto {chave}: {valor}")
```

```python
livros = [
  {
    "Título":"Diário de um Banana",
    "Autor":"Pinta",
    "Ano":2014   
  },  

  {
    "Título":"Clareou",
    "Autor":"Pixote",
    "Ano":2025   
  },

  {
    "Título":"O mar",
    "Autor":"Jesus",
    "Ano":1900   
  }             
]

print("Informações dos livros: ")

cont=1

for livro in livros:
  print(f"Título{cont}: {livro["Título"]}, Autor{cont}: {livro["Autor"]}, Ano{cont}: {livro["Ano"]}")
  cont+=1

livros.pop(0)

print("Livros Atualizados!!!")

cont=1
for livro in livros:
  print(f"Título{cont}: {livro["Título"]}, Autor{cont}: {livro["Autor"]}, Ano{cont}: {livro["Ano"]}")
  cont+=1

#E o comando para remover apenas um item de um dicionario dentro dessa lista como por exemplo o título apenas, o comando fica assim: livros[0].pop("Título"). 
```

```python
livros = []

# Definindo quantos livros o usuário quer adicionar
quantidade = int(input("Quantos livros deseja cadastrar? "))

for i in range(quantidade):
    print(f"\nCadastro do livro {i+1}:")
    titulo = input("Título: ")
    autor = input("Autor: ")
    ano = input("Ano: ")
    
    # Criando o dicionário e adicionando à lista
    novo_livro = {
        "Título": titulo,
        "Autor": autor,
        "Ano": ano
    }
    livros.append(novo_livro)

print("\nInformações dos livros: ")

cont=1
for livro in livros:
    print(f"Título{cont}: {livro['Título']}, Autor: {livro['Autor']}, Ano: {livro['Ano']}")
    cont+=1

livros.pop(0)

print("\nLivros Atualizados!!!")

cont=1
for livro in livros:
    print(f"Título{cont}: {livro['Título']}, Autor: {livro['Autor']}, Ano: {livro['Ano']}")
    cont+=1
```
